```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("training_dataset.csv")
```

```python
df.head()
```

```python
df["delta_start"] = np.sqrt(np.square(df["start_x"] - df["prev_x"]) + np.square(df["start_y"] - df["prev_y"]))
df["delta_end"] = np.sqrt(np.square(df["end_x"] - df["prev_x"]) + np.square(df["end_y"] - df["prev_y"]))
df.head()
```

```python
df.to_csv("training_dataset.csv",index=False)
```

```python
plt.plot(df["delta_start"][:100],df["prev_speed"][:100])
plt.show()
```
